```python
import pandas as pd

# ── Load datasets ────────────────────────────────────────────────
df_15k = pd.read_csv('Sleep_Data_Sampled.csv')   # replace filename if needed
df_5k = pd.read_csv('expanded_sleep_data_with_bio.csv')

# ── Remove unwanted sleep disorder categories ONLY from 5k ──────
df_5k = df_5k[
    ~df_5k['Sleep Disorder'].isin([
        'Narcolepsy',
        'REM Behavior Disorder'
    ])
]

# ── Remove biosignal columns from 5k dataset ────────────────────
biosignal_cols = [
    'EEG_Alpha',
    'EEG_Beta',
    'EEG_Theta',
    'ECG_HeartRate',
    'ECG_QRS_Duration',
    'ECF_Na',
    'ECF_K',
    'EMG_Mean',
    'EMG_Max'
]

# Remove Person ID column from BOTH datasets
df_15k = df_15k.drop(columns=['Person ID'])
df_5k = df_5k.drop(columns=['Person ID'])

# Remove Blood Pressure column from BOTH datasets
df_15k = df_15k.drop(columns=['Blood Pressure'])
df_5k = df_5k.drop(columns=['Blood Pressure'])
df_5k = df_5k.drop(columns=biosignal_cols)


# ── Combine datasets ─────────────────────────────────────────────
combined_df = pd.concat([df_15k, df_5k], ignore_index=True)

# ── Shuffle rows (recommended for ML) ────────────────────────────
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)

# ── Verify result ────────────────────────────────────────────────
print("Combined dataset shape:")
print(combined_df.shape)

print("\nColumns:")
print(combined_df.columns.tolist())

print("\nSleep Disorder classes:")
print(combined_df['Sleep Disorder'].value_counts())

# ── Save final dataset ───────────────────────────────────────────
combined_df.to_csv('combined_sleep_dataset.csv', index=False)

print("\nCombined dataset saved successfully.")
```
